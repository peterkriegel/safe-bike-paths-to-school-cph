```python
import pandas as pd
```

```python
import geopandas as gpd
```

```python
path = "../data/raw/Adresser til brug af skoledistrikter i KK 2026_02_19.xlsx"
```

```python
xls = pd.ExcelFile(path)
xls.sheet_names
```

```python
df = pd.read_excel(path, sheet_name=xls.sheet_names[0])
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df["Longitude"], df["Latitude"]),
    crs="EPSG:4326"
)

gdf.head()
```

```python
gdf = gdf.to_crs(epsg=25832)  # Denmark has the UTM zone 32
```
