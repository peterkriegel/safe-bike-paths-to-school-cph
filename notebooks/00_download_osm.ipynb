```python
import osmnx as ox
```

```python
place = "Copenhagen Municipality, Denmark"
G = ox.graph_from_place(place, network_type="bike")

ox.save_graphml(G, filepath="../data/cph_bike_network.graphml")
```

```python
G = ox.load_graphml("../data/cph_bike_network.graphml")

print(len(G.nodes), "nodes")
print(len(G.edges), "edges")
```

```python
edges = ox.graph_to_gdfs(G, nodes=False)
```
